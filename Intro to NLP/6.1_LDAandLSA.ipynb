```python
import pandas as pd
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
```

```python
#import gensim
#import gensim.corpora as corpora
#!pip install gensim
```

```python
data = pd.read_csv("./data_CSV/news_articles.csv")
data.head()
```

```python
#remove punctuation
data["remove_punctuation"] = data["content"].apply(lambda x: re.sub(r"([^\w\s])", "", x))
#remove Stopwords
en_stopwords = stopwords.words("english")
data["remove_stopwords"] = data["remove_punctuation"].apply(lambda x: " ".join([word for word in x.split() if word not in en_stopwords]))
#tokenize
data["tokenize"] = data["remove_stopwords"].apply(lambda x: word_tokenize(x))
#Stemming
ps = PorterStemmer()
data["stemming"] = data["tokenize"].apply(lambda tokens: [ps.stem(token) for token in tokens])
data.head()
```
